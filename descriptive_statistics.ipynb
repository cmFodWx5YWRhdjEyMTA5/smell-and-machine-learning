```python
import csv
import os
import shutil
import sys
import subprocess
import time
import git
from git import Repo
import pandas as pd
import numpy as np
from pydriller import RepositoryMining
from pydriller import GitRepository
import matplotlib.pyplot as plt
```

```python
PATH_REPOSITORIES = 'repositories'
PATH_RESULTS = 'results'
PATH_LOC = 'loc_results'
FILE_LIST_REPOSITORIES = 'filtered_data.csv'
FILE_DOWN = 'data/downloads.csv'
FILE_LOC = 'data/loc_repositories_total.csv'
FILE_SMELLS = 'data/smells_total.csv'
```

```python
#load csv repositories data
list_repositories = pd.read_csv(FILE_LIST_REPOSITORIES)
list_repositories
```

```python
#verificar quais repositórios fez download
dataDownload = []
for repository in list_repositories.sourceRep :
    rep_name = repository.split('https://github.com/',1)[1]
    name_path = rep_name.replace('/','_')
    path_rep_clone = PATH_REPOSITORIES + '/' + name_path
    if (os.path.exists(path_rep_clone)):
        gr = GitRepository(path_rep_clone)        
        dataDownload.append({'name' : rep_name, 'link' : repository, 'path' : path_rep_clone, 'commits' : gr.total_commits()})
# len(dataDownload)
# dataDownload 
```

```python
dfDownload = pd.DataFrame(dataDownload,index=None,columns=['name','link','path','commits'])
dfDownload
```

```python
dfDownload = pd.read_csv(FILE_DOWN)
count = 0
data_loc = []
for repository in dfDownload.link:
    rep_name = repository.split('https://github.com/',1)[1]
    name_path = rep_name.replace('/','_')
    path_rep_clone = PATH_REPOSITORIES + '/' + name_path
    path_name_repository = os.getcwd() + '/' + path_rep_clone
    path_result_loc = os.getcwd() + '/' + PATH_LOC + '/' + name_path + '.csv'   
    command_loc = 'cloc --csv {} --out="{}"'.format(path_name_repository,path_result_loc)
    #os.system(command_loc)
     #create file brank
#     print(path_result_loc)

    isSet = False
    
    if not (os.path.exists(path_result_loc)):
        isSet = True
        data_loc.append({'name' : rep_name, 'loc': 0,'cfiles': 0})
    else:
        loc_data = pd.read_csv(path_result_loc)
        for index, row in loc_data.iterrows():
            if row['language'] == 'Java':
                isSet = True
                data_loc.append({'name' : rep_name, 'loc': row['code'], 'files': row['files']})
    if not isSet:
        data_loc.append({'name' : rep_name, 'loc': 0, 'files': row['files']})
        
        
#         print('not exist')
#         with open(path_result_loc, 'w') as csvfile:
#             spamwriter = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
#             spamwriter.writerow(['files','language','blank','comment','code'])
#             spamwriter.writerows([0,'Java',0,0,0])
            
```

```python
downloads = pd.read_csv(FILE_DOWN)
downloads
```

```python
smells_repository = []
count = 0
for path in downloads.name:
    path_file_result  =  PATH_RESULTS + '/' + str(count) + '_' + path.replace('/','_') + '.csv'
    print(path_file_result)
    df = pd.read_csv(path_file_result)
    rules = df['Rule'].value_counts()
    
    total = getattr(rules,'TooManyMethods',0) + getattr(rules,'ExcessiveMethodLength',0) + getattr(rules,'ExcessiveClassLength',0) + getattr(rules,'DataClass',0) + getattr(rules,'ExcessiveParameterList',0)
    
    smells_repository.append({
       'name' : path,
       'tmm' : getattr(rules,'TooManyMethods',0), 
       'lm' : getattr(rules,'ExcessiveMethodLength',0), 
       'lc' : getattr(rules,'ExcessiveClassLength',0), 
       'dc' : getattr(rules,'DataClass',0), 
       'lpl' : getattr(rules,'ExcessiveParameterList',0),
       'smells' : total
    })
    count = count + 1
```

```python
dfSmells = pd.DataFrame(smells_repository,index=None,columns=['name','tmm','lm','lc','dc','lpl','smells'])
dfSmells
```

```python
dfSmells.to_csv('data/smells_total.csv')
```
