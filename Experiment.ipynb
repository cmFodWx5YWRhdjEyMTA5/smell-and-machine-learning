```python
import csv
import os
import shutil
import sys
import subprocess
import time
import git
from git import Repo
import pandas as pd
import numpy as np
from pydriller import RepositoryMining
from pydriller import GitRepository
import matplotlib.pyplot as plt
```

```python
PATH_REPOSITORIES = 'repositories'
PATH_RESULTS = 'results'
PATH_LOC = 'loc_results'
FILE_LIST_REPOSITORIES = 'filtered_data.csv'
PMD = '/home/matheus/pmd-bin-6.9.0/bin/run.sh pmd'
```

```python
#created path repositories 
if not os.path.exists(PATH_REPOSITORIES):
    os.makedirs(PATH_REPOSITORIES)

#created path results 
if not os.path.exists(PATH_RESULTS):
    os.makedirs(PATH_RESULTS)
```

```python
#load csv repositories data
list_repositories = pd.read_csv(FILE_LIST_REPOSITORIES)
# list_repositories
```

```python
for repository in list_repositories.sourceRep:
    rep_name = repository.split('https://github.com/',1)[1]
    name_path = rep_name.replace('/','_')
    path_rep_clone = PATH_REPOSITORIES + '/' + name_path
    
    if not (os.path.exists(path_rep_clone)): #clona se no existe
        try:
            print('cloning repostitory %s...' %(repository))
            Repo.clone_from(repository,path_rep_clone)    
        except:
            print('not cloning or repostitory exists %s...' %(repository))
    else:
        print('repository %s exist' %(repository))
print('clone completed...')
```

```python
#verificar quais repositrios fez download
dataDownload = []
for repository in list_repositories.sourceRep :
    rep_name = repository.split('https://github.com/',1)[1]
    name_path = rep_name.replace('/','_')
    path_rep_clone = PATH_REPOSITORIES + '/' + name_path
    if (os.path.exists(path_rep_clone)):
        gr = GitRepository(path_rep_clone)        
        dataDownload.append({'name' : rep_name, 'link' : repository, 'path' : path_rep_clone, 'commits' : gr.total_commits()})
# len(dataDownload)
# dataDownload 
```

```python
dfDownload = pd.DataFrame(dataDownload,index=None,columns=['name','link','path','commits'])
dfDownload.to_csv('data/downloads.csv')
```

```python
#monta o resultado final com todos as avaliaes do PMD
smells_repository = []
for index,row in dfDownload.iterrows():
    try:
        print('exec pmd in {}...'.format(row['name']))
        #pmd
        path_name_repository = os.getcwd() + '/' + row['path']
        path_name_result = os.getcwd() + '/results/' + str(index) + '_' + str(row['name']).replace('/','_') + '.csv'
        str_command = PMD + ' -d {} -f csv -R "category/java/design.xml/ExcessiveClassLength, category/java/design.xml/ExcessiveMethodLength, category/java/design.xml/ExcessiveParameterList, category/java/design.xml/DataClass, category/java/design.xml/TooManyMethods" -r {}'.format(path_name_repository,path_name_result)
        if not os.path.exists(path_name_result):
            os.system(str_command)

        #save data
        dfPMD = pd.read_csv(path_name_result)


        for inPMD, rwPMD in dfPMD.iterrows(): #percorre todos erros
            TMM = 1 if rwPMD['Rule'] == 'TooManyMethods' else 0
            LM = 1 if rwPMD['Rule'] == 'ExcessiveMethodLength' else 0
            LC = 1 if rwPMD['Rule'] == 'ExcessiveClassLength' else 0
            DC = 1 if rwPMD['Rule'] == 'DataClass' else 0
            LPL = 1 if rwPMD['Rule'] == 'ExcessiveParameterList' else 0

            smells_repository.append({
                'name' : row['name'],
                'link' : row['link'],
                'path' : row['path'],
                'file' : rwPMD['File'],
                'line' : rwPMD['Line'],

                'tmm' : TMM,
                'lm' : LM,
                'lc' : LC,
                'dc' : DC,
                'lpl' : LPL,
                'rule' : rwPMD['Rule']
            })
    except:
        print('no exec pmd...')
        
df1 = pd.DataFrame(smells_repository)
# dfPMD      
#df1.to_csv('teste.csv')
    
    
```

### Comea a fazer a separao de treino e teste aqui.

```python
resultFinal = pd.read_csv('data/result_finaly.csv')
resultFinal.head()
```

```python
values_counts = resultFinal['rule'].value_counts()
values_counts
```

```python
total = resultFinal['rule'].count()
total
```

### % de representao do dataset para cada Smell

```python
representation = values_counts/total
representation*100
```

```python
# Define a quantidade para ser treinada, que  80% da quantidade total com base na quantidade de cada smells
```

```python
train = values_counts*0.8
train
```

```python
# Define a quantidade para ser testada, que  o restante dos 80% da quantidade total, ou seja 20% com base na quantidade dos smells
test = values_counts-train
test
```

```python
# define a quantidade para treino que tenha smell, com base no % de representatividade em funo do geral
train_smell = train*representation
train_smell
```

```python
#define a quantidade de treino com artefatos sem smells
train_notsmell = train-train_smell
train_notsmell
```

```python
# define a quantidade para teste que tenha smell, com base no % de representatividade em funo do geral
test_smell = test*representation
test_smell
```

```python
#define a quantidade de test com artefatos sem smells
test_notsmell = test-test_smell
test_notsmell
```

```python
dataset = []
count_tmm = 0
count_lm = 0
count_lc = 0
count_dc = 0
count_lpl = 0

count_n_tmm = 0
count_n_lm = 0
count_n_lc = 0
count_n_dc = 0
count_n_lpl = 0

count_test_tmm = 0
count_test_lm = 0
count_test_lc = 0
count_test_dc = 0
count_test_lpl = 0

count_test_n_tmm = 0
count_test_n_lm = 0
count_test_n_lc = 0
count_test_n_dc = 0
count_test_n_lpl = 0

for idDF,rwDF in resultFinal.iterrows():
    
    #escolhe se  treino
    tmm_train = 0
    lm_train = 0
    lc_train = 0
    dc_train = 0
    lpl_train = 0
    
    #escolhe se  teste
    tmm_test = 0
    lm_test = 0
    lc_test = 0
    dc_test = 0
    lpl_test = 0
    
    #qual grupo faz parte
    tmm = 0
    lm = 0
    lc = 0
    dc = 0
    lpl = 0
 
    
    #treino para no nulos
    #TooManyMethods
    if count_tmm < int(train_smell['TooManyMethods']) and rwDF['tmm']: #avalia para treino
        tmm_train = 1
        tmm = 1
        count_tmm +=1
        
    elif count_test_tmm < int(test_smell['TooManyMethods']) and rwDF['tmm']: #avalia para teste
        tmm_test = 1
        tmm = 1
        count_test_tmm +=1
        
        
    
    #ExcessiveMethodLength
    if count_lm < int(train_smell['ExcessiveMethodLength']) and rwDF['lm']:
        lm_train = 1
        lm = 1
        count_lm +=1
    elif count_test_lm < int(test_smell['ExcessiveMethodLength']) and rwDF['lm']: #avalia para teste
        lm_test = 1
        lm = 1
        count_test_lm +=1
        
    #DataClass
    if count_dc < int(train_smell['DataClass']) and rwDF['dc']:
        dc_train = 1
        dc = 1
        count_dc +=1
    elif count_test_dc < int(test_smell['DataClass']) and rwDF['dc']: #avalia para teste
        dc_test = 1
        dc = 1
        count_test_dc +=1
        
    
    #ExcessiveClassLength
    if count_lc < int(train_smell['ExcessiveClassLength']) and rwDF['lc']:
        lc_train = 1
        lc = 1
        count_lc +=1
    elif count_test_dc < int(test_smell['ExcessiveClassLength']) and rwDF['lc']: #avalia para teste
        lc_test = 1
        lc = 1
        count_test_lc +=1
        
    #ExcessiveParameterList
    if count_lpl < int(train_smell['ExcessiveParameterList']) and rwDF['lpl']:
        lpl_train = 1
        lpl = 1
        count_lpl +=1
    elif count_test_dc < int(test_smell['ExcessiveParameterList']) and rwDF['lpl']: #avalia para teste
        lpl_test = 1
        lpl = 1
        count_test_lpl +=1
        
        
    #treino para nulos
    #treino para no nulos
    if count_n_tmm < int(train_notsmell['TooManyMethods']) and not rwDF['tmm'] and not tmm_train:
        tmm_train = 1
        tmm = 1
        count_n_tmm +=1
    elif count_test_n_tmm < int(test_notsmell['TooManyMethods']) and not rwDF['tmm'] and not tmm_train and not tmm_test:
        tmm_test = 1
        tmm = 1
        count_test_n_tmm +=1
        
        
    if count_n_lm < int(train_notsmell['ExcessiveMethodLength']) and not rwDF['lm'] and not lm_train:
        lm_train = 1
        lm = 1
        count_n_lm +=1
    elif count_test_n_lm < int(test_notsmell['ExcessiveMethodLength']) and not rwDF['lm'] and not lm_train and not lm_test:
        lm_test = 1
        lm = 1
        count_test_n_lm +=1
        
        
    if count_n_dc < int(train_notsmell['DataClass']) and not rwDF['dc'] and not dc_train:
        dc_train = 1
        dc = 1
        count_n_dc +=1
    elif count_test_n_dc < int(test_notsmell['DataClass']) and not rwDF['dc'] and not dc_train and not dc_test:
        dc_test = 1
        dc = 1
        count_test_n_dc +=1
        
        
    if count_n_lc < int(train_notsmell['ExcessiveClassLength']) and not rwDF['lc'] and not lc_train:
        lc_train = 1
        lc = 1
        count_n_lc +=1
    elif count_test_n_lc < int(test_notsmell['ExcessiveClassLength']) and not rwDF['lc'] and not lc_train and not lc_test:
        lc_test = 1
        lc = 1
        count_test_n_lc +=1
        
        
    if count_n_lpl < int(train_notsmell['ExcessiveParameterList']) and not rwDF['lpl'] and not lpl_train:
        lpl_train = 1
        lpl = 1
        count_n_lpl +=1
    elif count_test_n_lpl < int(test_notsmell['ExcessiveParameterList']) and not rwDF['lpl'] and not lpl_train and not lpl_test:
        lpl_test = 1
        lpl = 1
        count_test_n_lpl +=1
    
    
    
    dataset.append({
        'name' : rwDF['name'],
        'link' : rwDF['link'],
        'path' : rwDF['path'],
        'file' : rwDF['file'],
        'line' : rwDF['line'],
        
        'tmm' : tmm,
        'has_tmm' : rwDF['tmm'],
        'train_tmm' : tmm_train,
        'test_tmm' : tmm_test,
        
        'lm' : lm,
        'has_lm' : rwDF['lm'],
        'train_lm' : lm_train,
        'test_lm' : lm_test,
        
        'lc' : lc,
        'has_lc' : rwDF['lc'],
        'train_lc' : lc_train,
        'test_lc' : lc_test,
        
        'dc' : dc,
        'has_dc' : rwDF['dc'],
        'train_dc' : dc_train,
        'test_dc' : dc_test,
        
        'lpl' : lpl,
        'has_lpl' : rwDF['lpl'],
        'train_lpl' : lpl_train,
        'test_lpl' : lpl_test,
        
        'rule' : rwDF['rule']
    })
    
data = pd.DataFrame(dataset,index=None,columns=['name', 'link', 'path', 'file', 'line', 
                                                'tmm', 'has_tmm', 'train_tmm', 'test_tmm', 
                                                'lm', 'has_lm', 'train_lm', 'test_lm', 
                                                'lc', 'has_lc', 'train_lc', 'test_lc', 
                                                'dc', 'has_dc', 'train_dc', 'test_dc', 
                                                'lpl', 'has_lpl', 'train_lpl', 'test_lpl', 
                                                'rule'])

    
```
