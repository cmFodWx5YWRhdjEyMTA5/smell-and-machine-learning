# Download repositories and code analyzer with PMD

The purpose of this script is to download the repositories and labeling some smells using PMD.

## Import dependencies

```python
import csv
import os
import shutil
import sys
import subprocess
import time
import git
import pandas as pd
import numpy as np
from pydriller import RepositoryMining
from pydriller import GitRepository
import matplotlib.pyplot as plt
```

## Declaration of constants

```python
PATH_REPOSITORIES = 'repositories'
PATH_RESULTS = 'results'
FILE_LIST_REPOSITORIES = 'filtered_data.csv'
```

```python
#created path repositories 
if not os.path.exists(PATH_REPOSITORIES):
    os.makedirs(PATH_REPOSITORIES)

#created path results 
if not os.path.exists(PATH_RESULTS):
    os.makedirs(PATH_RESULTS)
```

```python
#load csv repositories data
list_repositories = pd.read_csv(FILE_LIST_REPOSITORIES)
# list_repositories
```

```python
for repository in list_repositories.sourceRep :
#     print('cloning repostitory %s...' %(repository))
    
    rep_name = repository.split('https://github.com/',1)[1]
    path_rep_clone = PATH_REPOSITORIES + '/' + rep_name.split('/',1)[1]
      
    
    if not (os.path.exists(path_rep_clone)): #clona se n√£o existe
        try:
            print('cloning repostitory %s...' %(repository))
            #git.Git(PATH_REPOSITORIES).clone(repository)    
        except:
            print('not cloning repostitory %s...' %(repository))
    else:
        print('repository %s exist' %(repository))
print('clolne completed...')
```

```python
#verificar quais reposit√≥rios fez download
dataDownload = []
for repository in list_repositories.sourceRep :
    rep_name = repository.split('https://github.com/',1)[1]
    path_rep_clone = PATH_REPOSITORIES + '/' + rep_name.split('/',1)[1]
    if (os.path.exists(path_rep_clone)):
        gr = GitRepository(path_rep_clone)        
        dataDownload.append([rep_name,repository,path_rep_clone,gr.total_commits()])
# len(dataDownload)
# dataDownload        
```

```python
# c = csv.writer(open('rep_down.csv','wb'))
# # for row in dataDownload:

# c.writerows(dataDownload)

with open('rep_down.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['name','link','path','commits'])
    spamwriter.writerows(dataDownload)
    

```
